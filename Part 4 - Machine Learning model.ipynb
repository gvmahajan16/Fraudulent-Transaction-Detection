```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import  MinMaxScaler,RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler,OrdinalEncoder, PowerTransformer


numerical_transformer = Pipeline(steps=[
    
    ('imputer',SimpleImputer(strategy='median')),
    ('scaler', RobustScaler()),
    ('transformer', PowerTransformer())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ohe',OneHotEncoder(handle_unknown='ignore'))
])

ordinal_transformer=Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ord_enc',OrdinalEncoder())
    
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numeric_column),
        ('cat', categorical_transformer, oh_columns),
        ('cat_or',ordinal_transformer,or_columns)
    ]
)
```

```python
from sklearn.model_selection import  train_test_split
# separate dataset into train and test
x_train, x_test, y_train, y_test = train_test_split(X_final,y_final,test_size=0.2,random_state=42)
```

```python
def create_model_lg():
    from sklearn.linear_model import LogisticRegressionCV
    model = LogisticRegressionCV()
    model.fit(x_train, y_train)
    return model
```

```python
def create_model_knn():
    from sklearn.neighbors import KNeighborsClassifier
    model = KNeighborsClassifier(n_neighbors=5)
    model.fit(x_train, y_train)
    return model
```

```python
def create_model_dt():
    from sklearn.tree import DecisionTreeClassifier
    model = DecisionTreeClassifier(max_depth=5)
    model.fit(x_train, y_train)
    return model
```

```python
def create_model_rf():
    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier(n_estimators=20, max_depth=5)
    model.fit(x_train, y_train)
    return model
```

```python
def create_model_xgb():
    from xgboost import XGBClassifier
    model = XGBClassifier()
    model.fit(x_train, y_train)
    return model
```

```python
model_lg = create_model_lg()
model_knn = create_model_knn()
# model_svm = create_model_svm()
model_dt = create_model_dt()
model_rf = create_model_rf()
# model_gb = create_model_gb()
model_xgb = create_model_xgb()
# model_ada = create_model_ada()
```

```python
def evaluate_model(model):
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
    
    y_true = y_test
    if model == model_knn:
        y_pred = model.predict(x_test)
    else:
        y_pred = model.predict(x_test)
    
    accuracy = f"{accuracy_score(y_true, y_pred):0.2f}"
    precision = f"{precision_score(y_true, y_pred):0.2f}"
    recall = f"{recall_score(y_true, y_pred):0.2f}"
    f1 = f"{f1_score(y_true, y_pred):0.2f}"
    roc = f"{roc_auc_score(y_true, y_pred):0.2f}"
    
    return accuracy, precision, recall, f1 , roc
```

```python
def evaluate_model_train(model):
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
    
    y_true = y_train
    if model == model_knn:
        y_pred = model.predict(x_train)
    else:
        y_pred = model.predict(x_train)
    
    accuracy = f"{accuracy_score(y_true, y_pred):0.2f}"
    precision = f"{precision_score(y_true, y_pred):0.2f}"
    recall = f"{recall_score(y_true, y_pred):0.2f}"
    f1 = f"{f1_score(y_true, y_pred):0.2f}"
    roc = f"{roc_auc_score(y_true, y_pred):0.2f}"
    
    return accuracy, precision, recall, f1 , roc
```

```python
models = [
    {"name": "LG", "model": model_lg},
    {"name": "KNN", "model": model_knn},
    {"name": "DT", "model": model_dt},
    {"name": "RF", "model": model_rf},
    {"name": "XGB", "model": model_xgb},
]

print('-' * 72)
print(f"|{'Algo':^10}|{'Accuracy':^15}|{'Precision':^15}|{'Recall':^15}|{'F1':^10}|{'ROC':^10}|")
print('-' * 72)
for model_info in models:
    evaluation = evaluate_model_train(model_info['model'])
    print(f"|{model_info['name']:^10}|{evaluation[0]:^15}|{evaluation[1]:^15}|{evaluation[2]:^15}|{evaluation[3]:^10}|{evaluation[4]:^10}|")
    
print('-' * 72)
```
